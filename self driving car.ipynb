```python
# Import necessary libraries for data handling and training
import torch
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np

# Dataset class for YOLOv10 (assumes YOLO formatted labels)
class SelfDrivingDataset(Dataset):
    def __init__(self, images_dir, labels_dir, transform=None):
        self.images_dir = images_dir
        self.labels_dir = labels_dir
        self.transform = transform
        self.image_files = os.listdir(images_dir)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.images_dir, self.image_files[idx])
        label_path = os.path.join(self.labels_dir, self.image_files[idx].replace('.jpg', '.txt'))

        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        with open(label_path, 'r') as f:
            labels = np.array([list(map(float, line.split())) for line in f])

        if self.transform:
            transformed = self.transform(image=image, bboxes=labels[:, 1:], category_ids=labels[:, 0])
            image = transformed['image']
            labels = np.hstack((transformed['category_ids'][:, None], transformed['bboxes']))

        return image, labels
```

```python
import torch
from torch.utils.data import DataLoader
from ultralytics import YOLO

# Load the YOLOv10 model
model = YOLO("/content/yolov10/weights/yolov10x.pt")
  # Replace with your model file path

# Set device to CPU
device = torch.device('cpu')
model = model.to(device)
dataset="../data.yaml"
# Assume you have a DataLoader for your dataset
train_loader = DataLoader(dataset, batch_size=16, shuffle=True)

# Optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = torch.nn.CrossEntropyLoss()

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0
    for images, labels in train_loader:
        # Move data to CPU
        images = images.to(device)
        labels = labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader)}')
# Train the model
model.detect(data="./data.yaml", epochs=5, imgsz=512, batch=16)

# Validate the model on the validation set
metrics = model.val()

# Save the model weights after training
model.save_weights("/content/yolov10/weights/yolov10x.pt")

# Print the validation metrics
print("C:/Users/engba/Documents/yolov10/Validation Metrics: ", metrics)
```
